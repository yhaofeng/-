```python
def three_sigma(df):
    mean = df.mean()
    std = df.std()
    sigma_up = mean + 3*std
    sigma_low = mean - 3*std
    index = []
    j = 0
    for i in df.values:
        if i<sigma_low or sigma_up<i:
            index.append(j)
        j = j+1
    return index
```
